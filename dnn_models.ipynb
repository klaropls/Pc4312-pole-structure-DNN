```python
import chainer
import chainer.links as L
import chainer.functions as F
```

```python
class MLP1L(chainer.Chain):
    #Two hidden layers with 100-100 nodes
    def __init__(self):
        super(MLP1L, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(200, 200, nobias=False, initialW=None, initial_bias=None)
            self.l2 = L.Linear(None, 200, nobias=False, initialW=None, initial_bias=None)
            self.l3 = L.Linear(None, 8, nobias=False, initialW=None, initial_bias=None)

    def forward(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        return self.l3(h2)

###First SMORMS 200-100
###Second SMORMS 300-300-300

# class MLP2L(chainer.Chain):
#     #Two hidden layers with 200-200 nodes
#     def __init__(self):
#         super(MLP2L, self).__init__()
#         with self.init_scope():
#             self.l1 = L.Linear(None, 200, nobias=False, initialW=None, initial_bias=None)
#             self.l2 = L.Linear(None, 200, nobias=False, initialW=None, initial_bias=None)
#             self.l3 = L.Linear(None, 35, nobias=False, initialW=None, initial_bias=None)

#     def forward(self, x):
#         h1 = F.relu(self.l1(x))
#         h2 = F.relu(self.l2(h1))
#         return self.l3(h2)
    
# class MLP3L(chainer.Chain):
#     #Two hidden layers with 300-300 nodes
#     def __init__(self):
#         super(MLP2L, self).__init__()
#         with self.init_scope():
#             self.l1 = L.Linear(None, 300, nobias=False, initialW=None, initial_bias=None)
#             self.l2 = L.Linear(None, 300, nobias=False, initialW=None, initial_bias=None)
#             self.l3 = L.Linear(None, 35, nobias=False, initialW=None, initial_bias=None)

#     def forward(self, x):
#         h1 = F.relu(self.l1(x))
#         h2 = F.relu(self.l2(h1))
#         return self.l3(h2)

class MLP2L(chainer.Chain):
    #Three hidden layers with 200-200-200 nodes
    def __init__(self):
        super(MLP2L, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(None, 250, nobias=False, initialW=None, initial_bias=None)
            self.l2 = L.Linear(None, 100, nobias=False, initialW=None, initial_bias=None)
            self.l3 = L.Linear(None,  50, nobias=False, initialW=None, initial_bias=None)
            self.l4 = L.Linear(None, 35, nobias=False, initialW=None, initial_bias=None)

    def forward(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        h3 = F.relu(self.l3(h2))
        return self.l4(h3)

class MLP3L(chainer.Chain):
    #Three hidden layers with 250-250-250 nodes
    def __init__(self):
        super(MLP3L, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(200, 300, nobias=False, initialW=None, initial_bias=None)
            self.l2 = L.Linear(None, 300, nobias=False, initialW=None, initial_bias=None)
            self.l3 = L.Linear(None, 300, nobias=False, initialW=None, initial_bias=None)
            self.l4 = L.Linear(None, 8, nobias=False, initialW=None, initial_bias=None)

    def forward(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        h3 = F.relu(self.l3(h2))
        return self.l4(h3)    
    
class MLP4L(chainer.Chain):
    #Three hidden layers with 250-250-250 nodes
    def __init__(self):
        super(MLP4L, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(200, 100, nobias=False, initialW=None, initial_bias=None)
            self.l2 = L.Linear(None, 100, nobias=False, initialW=None, initial_bias=None)
            self.l3 = L.Linear(None, 100, nobias=False, initialW=None, initial_bias=None)
            self.l4 = L.Linear(None, 8, nobias=False, initialW=None, initial_bias=None)

    def forward(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        h3 = F.relu(self.l3(h2))
        return self.l4(h3)  

class MLP5L(chainer.Chain):
    #Two hidden layers with 200-100 nodes
    def __init__(self):
        super(MLP5L, self).__init__()
        with self.init_scope():
            self.l1 = L.Linear(200, 100, nobias=False, initialW=None, initial_bias=None)
            self.l2 = L.Linear(None, 100, nobias=False, initialW=None, initial_bias=None)
            self.l3 = L.Linear(None, 8, nobias=False, initialW=None, initial_bias=None)

    def forward(self, x):
        h1 = F.relu(self.l1(x))
        h2 = F.relu(self.l2(h1))
        return self.l3(h2)

num_arch = 5
```
